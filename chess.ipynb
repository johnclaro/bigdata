```python
import pandas as pd
data = pd.read_csv('files/openings.csv')
data = data[:10].iloc[::-1]
print(data)
```

```python
from matplotlib import pyplot as plot

data.plot(kind='barh', title='Top Openings', x='opening')
plot.title('Top Openings')
plot.ylabel('Openings')
plot.xlabel('Frequency')
```
