```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams["figure.figsize"] = (20,10)

df = pd.read_csv('files/plies.csv')

plt.plot('Plies', 'BlackWins', data=df, marker='', color='skyblue', linewidth=1, label='Black Wins')
plt.plot('Plies', 'WhiteWins', data=df, marker='', color='orange', linewidth=1, label='White Wins')
plt.plot('Plies', 'Draw', data=df, marker='', color='green', linewidth=1, label="Draw")
plt.legend()
plt.show()
```
